## Feature Engineering

```python
df['study_efficiency'] = df['gpa'] / (df['study_time_weekly'] + 1)

df['risk_score'] = (
    (df['absences'] / df['absences'].max()) * 0.6 +
    (df['gpa'].max() - df['gpa']) / df['gpa'].max() * 0.4
)

df['risk_level'] = pd.cut(
    df['risk_score'],
    bins=[0,0.33,0.66,1],
    labels=['Low','Medium','High']
)
```
